```python
import numpy as np
import math
from pettingzoo.mpe import simple_tag_v3

from CoordinatingAgent import CoordinatingAgent
from RandomAgent import RandomAgent

NUM_GOOD = 1
NUM_ADVERSARIES = 4
NUM_LANDMARKS = 0
MAX_CYCLES = 200

THRESHOLD = 1e-5
DISTANCE_THRESHOLD = 0.5
TARGET_VECT_THRESHOLD = 5

env = simple_tag_v3.parallel_env(num_good=NUM_GOOD, num_adversaries=NUM_ADVERSARIES, num_obstacles=NUM_LANDMARKS, max_cycles=MAX_CYCLES, continuous_actions=False, render_mode="human")
observations, infos = env.reset()

coord_agents = [CoordinatingAgent(name, NUM_ADVERSARIES, NUM_LANDMARKS) for name in env.agents if name != "agent_0"]
coord_agents.append(RandomAgent("agent_0", NUM_ADVERSARIES, NUM_LANDMARKS))
for agent in coord_agents:
        agent.observations = observations[agent.name]
        agent.update_observed_agent_positions()

# this implementation assumes one prey only
while env.agents:
    actions = {}
    sum_of_distances_from_target = 0
    for agent in coord_agents:
        actions[agent.name] = agent.get_action()
    observations, rewards, terminations, truncations, infos = env.step(actions)
    for agent in coord_agents:
        agent.observations = observations[agent.name]
        agent.update_observed_agent_positions()

    # print("rewards: ", rewards)

env.close()







```
